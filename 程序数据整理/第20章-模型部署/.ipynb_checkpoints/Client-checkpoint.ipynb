```python
from tensorflow_serving.apis import predict_pb2
from tensorflow_serving.apis import prediction_service_pb2_grpc
import grpc

def request_server(img_resized, server_url):
    '''
    用于向TensorFlow Serving服务请求推理结果的函数。
    :param img_resized: 经过预处理的待推理图片数组，numpy array，shape：(h, w, 3)
    :param server_url: TensorFlow Serving的地址加端口，str，如：'0.0.0.0:8500' 
    :return: 模型返回的结果数组，numpy array
    '''
    # Request.
    channel = grpc.insecure_channel(server_url)
    stub = prediction_service_pb2_grpc.PredictionServiceStub(channel)
    request = predict_pb2.PredictRequest()
    request.model_spec.name = "my_model"  # 模型名称，启动容器命令的model_name参数
    request.model_spec.signature_name = "serving_default"  # 签名名称，刚才叫你记下来的
    # "flatten_input"是你导出模型时设置的输入名称，刚才叫你记下来的
#     request.inputs["flatten_input"].CopyFrom(
#         tf.make_tensor_proto(img_resized, shape=[1, ] + list(img_resized.shape)))
    request.inputs["flatten_input"].CopyFrom(tf.make_tensor_proto(img_resized, dtype=tf.float32, shape=[1, ] + list(img_resized.shape)))
    # 5 secs timeout
    response = stub.Predict(request, 5.0)  
    # dense为输出名称，刚才叫你记下来的
    return np.asarray(response.outputs["dense"].float_val) 
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# 载入数据集
mnist = tf.keras.datasets.mnist
# 载入数据，数据载入的时候就已经划分好训练集和测试集
(x_train, y_train), (x_test, y_test) = mnist.load_data()
# 对训练集和测试集的数据进行归一化处理，有助于提升模型训练速度
x_train, x_test = x_train / 255.0, x_test / 255.0
plt.imshow(x_train[0],cmap='gray')
plt.show()
```

```python
# grpc地址及端口，为你容器所在机器的ip + 容器启动命令里面设置的port
server_url = '0.0.0.0:8500'
pre = request_server(x_train[0], server_url)
print("预测结果为：",np.argmax(pre))
```
